```python
# Library as needed
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd
```

```python
# Read BBC news file
file_path_bbc = "bbc_news.csv"
data_bbc = pd.read_csv(file_path_bbc)
df_bbc = data_bbc
df_bbc
```

##### Here we create a function to read Url in the csv file, download HTML content of the news and finally extract the title. We understand the format of the HTML file by multiple test before (i.e we only collect the Url have the same format, that is, the 'contents' class is stored in the first script card and the title is always the first content in the 'text' class). 
##### After finish collecting all titles, we store it in a dataframe and save to a new csv file.

```python
# Function to get all contents of the BBC news
def scrape_content_bbc(url_bbc):

    # Import HTML source through url
    response_bbc = requests.get(url_bbc)
    soup_bbc = BeautifulSoup(response_bbc.content, "html.parser")

    # Contents are stored in the first script card
    script_bbc = soup_bbc.find("script", type="application/json")
    
    # Access to the "contents" class of the JSON structure
    bbc_data = json.loads(script_bbc.text)
    url_bbc_id = url_bbc.split("/")[-1] # News are marked by using id
    bbc_content = bbc_data["props"]["pageProps"]["page"]['@"news","{}",'.format(url_bbc_id)]["contents"]

    # Extract all text part of the contents
    i = 0
    title_bbc = []
    for text_content in bbc_content:
        try:
            text = bbc_content[i]['model']['blocks'][0]['model']['text']
            title_bbc.append(text)
        except KeyError:
            # Handle the KeyError
            text = ""
        i += 1    
        # Just need the title for analysis
        if len(title_bbc) == 1:
            break
    
    completed_title_bbc = " ".join(title_bbc)
    # Return a completed title in a sentence
    return completed_title_bbc
```

```python
# Empty list to store all content
title_bbc_list = []

# Iterate through each row in the CSV
for index, row in df_bbc.iterrows():
    url = row['Url'] # Urls are stored in 'Url' column of csv file 
    title = scrape_content_bbc(url)
    title_bbc_list.append(title)
```

```python
# Add contents to dataframe
df_bbc['Title'] = title_bbc_list 
df_bbc
```
